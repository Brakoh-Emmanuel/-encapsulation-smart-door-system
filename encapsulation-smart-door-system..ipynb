```python
class Staff:
    def __init__(self, s_name, access_code):
        self.s_name = s_name
        self.__access_code = access_code  # private attribute

    # Property to view the access code (read-only)
    @property
    def access_code(self):
        # Authorized access logic could be added here
        return self.__access_code

    # Setter to update the access code with validation
    @access_code.setter
    def access_code(self, new_code):
        if len(new_code) >= 6:  # Example validation: minimum length of 6
            self.__access_code = new_code
            print("Access code updated successfully.")
        else:
            print("Error: Access code must be at least 6 characters long.")

    # Method to display staff information
    def display_info(self):
        print(f"Staff Name: {self.s_name}")
        print(f"Access Code: {self.__access_code}")
# Create staff member
staff1 = Staff("Alice Mensah", "abc123")

# Display staff info
staff1.display_info()

# View access code
print("Current Access Code:", staff1.access_code)

# Update access code
staff1.access_code = "newpass123"   # Valid update
staff1.access_code = "123"          # Invalid update (too short)

# Display updated info
staff1.display_info()

```
